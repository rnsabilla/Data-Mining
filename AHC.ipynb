```python
from matplotlib import pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
import numpy as np
```

```python
x = np.array([[2,4,4,3],[3,4,3,5],[4,3,2,5],[1,5,4,2],[3,2,1,3]])
```

```python
print(x)
```

```python
Z = linkage(x)
```

```python
print(Z)
```

```python
plt.figure(figsize=(25,10))
plt.title('Dendrogram Hierarchical Klastering')
plt.xlabel('data')
plt.ylabel('jarak')
dendrogram(Z)
plt.show()
```
